# Final Product: Do QB Transfers Pay Off?

**An Analysis of NCAA Transfer Portal Performance (2021-2024)**

## The Infographic

![QB Transfer ROI Infographic](../data/infographic.png)

## Description

I analyzed 147 college football quarterbacks who transferred between 2021-2024, comparing their performance before and after the transfer. The infographic combines three visualizations:

1. **Density Plot** - Shows the overall distribution of passer ratings before vs after transfer, showing a shorter left tail and longer right tail post-transfer, indicating QBs were more likely to improve significantly than decline significantly.

2. **Success Rate Bar Chart** - Breaks down improvement rates by transfer direction (UP/Lateral/DOWN), showing that 62.5% of QBs who transferred to better teams improved their passer rating

3. **Volume Metrics Comparison** - Compares passing yards and TD changes for QBs who transferred up vs down in team strength, showing that volume stats increase for those moving to worse programs and decrease for those moving to better ones.

### Key Takeaway

**Transferring to better programs helps QB efficiency, but on average, it comes at the cost of volume and big stats.** QBs who moved to teams with 5+ higher SP+ ratings saw an average +6.72 passer rating improvement, but 1363 **less** yards, 6.8 **less** TDs Those who transferred down saw a worse passer rating of 3.92 on average and **improved** by 540 yards, 3.2 TDs on average.
